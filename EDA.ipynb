```python
import gzip

communities = []

with gzip.open('/content/drive/MyDrive/RL/data/new/all_com.txt.gz', 'rt') as f:
    for line in f:
        nodes = list(map(int, line.strip().split()))
        communities.append(nodes)

print("Total communities:", len(communities))
```

```python
sizes = [len(c) for c in communities]

print("Min size:", min(sizes))
print("Max size:", max(sizes))
print("Average size:", sum(sizes)/len(sizes))
```

```python
import matplotlib.pyplot as plt

plt.hist(sizes, bins=50)
plt.xlabel("Community Size")
plt.ylabel("Frequency")
plt.title("Distribution of Community Sizes")
plt.show()
```

```python
sorted_sizes = sorted(sizes, reverse=True)

print("Top 10 largest communities:", sorted_sizes[:10])
```

```python
from collections import Counter

node_count = Counter()

for community in communities:
    for node in community:
        node_count[node] += 1

# how many communities each node belongs to
membership_counts = list(node_count.values())

print("Max memberships of a node:", max(membership_counts))
print("Average memberships:", sum(membership_counts)/len(membership_counts))
```

```python
plt.hist(membership_counts, bins=50)
plt.xlabel("Number of communities per node")
plt.ylabel("Frequency")
plt.title("Node Membership Distribution")
plt.show()
```
